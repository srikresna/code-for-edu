# Bagian 1

### Berdasarkan "height"

```python
no1 = pd.read_csv('pop1.csv')
no1
```

```python
# a. Carilah nilai mean, varians, dan simpangan bakunya.
mean = no1['height'].mean()
var = no1['height'].var()
std = no1['height'].std()

print('Mean:', mean)
print('Variance:', var)
print('Standard Deviation:', std)
```

```python
no1.describe()
```

```python
# b. Carilah Q1, Q2, dan Q3 (kalkulasi interpolasi dengan “midpoint” atau nilai tengah). 
Q1 = no1['height'].quantile(0.25)
Q2 = no1['height'].quantile(0.5)
Q3 = no1['height'].quantile(0.75)

print('Q1:', Q1)
print('Q2:', Q2)
print('Q3:', Q3)
```

```python
# c.	Apakah terdapat outlier? Jika ada, sebutkan jumlahnya! 
Q1 = no1['height'].quantile(0.25)
Q3 = no1['height'].quantile(0.75)
IQR = Q3 - Q1

print('IQR dari data adalah:', IQR)
print('Total data yang ada:', len(no1))
print('========================================')
outliers = no1[(no1['height'] < (Q1 - 1.5 * IQR)) | (no1['height'] > (Q3 + 1.5 * IQR))]
print('Outliers:', len(outliers))
```

### Berdasarkan data

```python
import matplotlib.pyplot as plt
```

```python
# a.	Boxplot dari “height” (jenis interpolasi bebas, dapat menggunakan linier atau nilai tengah)

# interpolasi menggunakan nilai tengah
plt.boxplot(no1['height'], vert=False, showmeans=True)
plt.title('Boxplot of height')
plt.show()
```

```python
# b. Histogram dari “height” dengan jumlah kelompok (bins) adalah 100. 

plt.hist(no1['height'], bins=100)
plt.title('Histogram of height')
plt.show()
```

```python
no2.info()
```

```python

# Index(['Month', 'DATE', 'ORDER', 'WH', 'ORDER\nNO', 'VENDOR\nNO',
#        'VENDOR NAME', 'SKU', 'SKU  DESCRIPTION', 'ORDER\nQTY ',
#        'RECEIVED\nQTY ', 'STATUS '],
#       dtype='object')



# a.	Carilah 5 produk SKU DESCRIPTION dengan jumlah order terbesar 
no2.groupby('SKU  DESCRIPTION')['ORDER\nQTY '].sum().sort_values(ascending=False).head(5)
```

```python
# b.	Berdasarkan poin a, buatlah diagram batang (bar chart) dari produk-produk tersebut 
top5 = no2.groupby('SKU  DESCRIPTION')['ORDER\nQTY '].sum().sort_values(ascending=False).head(5)
plt.bar(top5.index, top5.values)
plt.xticks(rotation=90)
plt.title('Top 5 SKU DESCRIPTION')
plt.show()
```

```python
# c.	Hitunglah nilai mean, median, Q1, Q3, dan IQR dari data-data poin a
mean = no2['ORDER\nQTY '].mean()
median = no2['ORDER\nQTY '].median()
Q1 = no2['ORDER\nQTY '].quantile(0.25)
Q3 = no2['ORDER\nQTY '].quantile(0.75)
IQR = Q3 - Q1

print('Mean:', mean)
print('Median:', median)
print('Q1:', Q1)
print('Q3:', Q3)
print('IQR:', IQR)
```

```python
# d.	Hitunglah nilai variasi (variance) dan simpangan baku (standard deviation) dari data-data poin a
var = no2['ORDER\nQTY '].var()
std = no2['ORDER\nQTY '].std()

print('Variance:', var)
print('Standard Deviation:', std)
```

```python
# total produk unik
top5 = no2.groupby('SKU  DESCRIPTION')['ORDER\nQTY '].sum().sort_values(ascending=False).head(5)
top5
```

```python
# e.	Carilah distribusi order produk pada poin a berdasarkan bulan. Gambarkan diagram batangnya, kemudian simpulkan apakah order produk tersebut menunjukkan bentuk kemiringan. Jika iya, miring ke arah mana. 
# parse terlebih dahulu DATE
no2['DATE'] = pd.to_datetime(no2['DATE'])
no2['Month'] = no2['DATE'].dt.month

# # SKU  DESCRIPTION RANK 1-5
# ATTACK PLUS SOFTENER 800 GR            159264
# LAURIER RELAX NIGHT WING 35CM 12'S      43944
# BIORE MEN'S DOUBLE SCRUB 100 GR         34656
# BIORE BF WHITENING SCRUB PCH 450 ML     32700
# BIORE BF PURE MILD PCH 450 ML           29400

# # produk ATTACK PLUS SOFTENER 800 GR histogram
# attack = no2[no2['SKU  DESCRIPTION'] == 'ATTACK PLUS SOFTENER 800 GR']
# attack = attack.groupby('Month')['ORDER\nQTY '].sum()
# plt.bar(attack.index, attack.values)
# plt.title('ATTACK PLUS SOFTENER 800 GR')
# plt.show()


fig, ax = plt.subplots(1, 5, figsize=(20, 5))
for i, sku in enumerate(top5.index):
    data = no2[no2['SKU  DESCRIPTION'] == sku]
    data = data.groupby('Month')['ORDER\nQTY '].sum()
    ax[i].bar(data.index, data.values)
    ax[i].set_title(sku)
plt.show()

# histogram
fig, ax = plt.subplots(1, 5, figsize=(20, 5))
for i, sku in enumerate(top5.index):
    data = no2[no2['SKU  DESCRIPTION'] == sku]
    ax[i].hist(data['ORDER\nQTY '], bins=20)
    ax[i].set_title(sku)
plt.show()
```
