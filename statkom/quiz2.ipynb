```python
# Sebuah perusahaan telekomunikasi mengklaim bahwa pelanggannya memiliki tagihan rata-rata
# sebesar $400 dengan simpangan baku $25 untuk menggunakan layanannya. Dilakukan pengambilan sampel sebanyak 50 pelanggan, 
# didapatkan rata-rata tagihannya adalah $250 dengan simpangan baku $15. 
# Berdasarkan informasi tersebut, apakah benar klaim yang dilakukan oleh perusahaan? (Gunakan alpha 0,05) (25 poin)

# H0 : μ = 400
# H1 : μ ≠ 400

# α = 0.05
# n = 50
# σ = 25
# x̄ = 250
# σ = 15

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Diketahui beberapa nilai yang diperlukan adalah sebagai berikut
n = 50 # jumlah sampel
σp = 25 # simpangan baku populasi
x̄ = 250 # rata-rata sampel
μ = 400 # rata-rata populasi
σs = 15 # simpangan baku sampel
α = 0.05 # tingkat signifikansi

# Menghitung z-score karena n > 30
z = (x̄ - μ) / (σp / np.sqrt(n))
print("z-score:", z)

# Menghitung z-critical
z_critical = stats.norm.ppf(1 - α/2)
print("z-critical:", z_critical)

# Menentukan keputusan
if np.abs(z) > z_critical:
    print("Tolak H0")
else:
    print("Gagal tolak H0")

# Menghitung p-value dengan two-tailed test karena H1: μ ≠ 400 
p = 2 * (1 - stats.norm.cdf(z))
print("p-value:", p)


print("\n")

# SOAL 2
# Jasmine baru memulai pekerjaan baru di Sales Force, sebuah perusahaan yang sangat 
# kompetitif. Pada data 16 sampel penjualan (sales) via telepon yang dilakukan oleh 
# Jasmine didapatkan bahwa terjadi penjualan sukses dengan rata-rata $108 dengan simpangan
# baku 5%. Lakukan pengujian dengan level signifikasi 5% dengan rata-rata penjualan
# sukses pada populasi paling tidak $100 dengan alternatifnya adalah kurang dari $100.
# Sesuai dengan peraturan perusahaan, setiap pegawai sales baru harus mampu menghasilkan 
# rata-rata penjualan sukses minimal $100 pada tahap percobaan. Berdasarkan pengujian,
# apakah kita dapat menyimpulkan bahwa Jasmine dapat memenuhi target masa percobaan? (25 poin)

# H0 : μ = 100
# H1 : μ < 100

# α = 0.05
# n = 16
# σ = 5
# x̄ = 108
# μ = 100

# Diketahui beberapa nilai yang diperlukan adalah sebagai berikut
n = 16 # jumlah sampel
σp = 5 # simpangan baku populasi
x̄ = 108 # rata-rata sampel
μ = 100 # rata-rata populasi
α = 0.05 # tingkat signifikansi

# Menghitung t-score karena n < 30
t = (x̄ - μ) / (σp / np.sqrt(n))
print("t-score:", t)

# Menghitung t-critical
t_critical = stats.t.ppf(α, n-1)
print("t-critical:", t_critical)

# Menentukan keputusan
if t < t_critical:
    print("Tolak H0")
else:
    print("Gagal tolak H0")

# Menghitung p-value dengan one-tailed test karena H1: μ < 100
p = stats.t.cdf(t, n-1)
print("p-value:", p)

print("\n")

# SOAL 3
# Berdasarkan survei yang dilakukan oleh Lembaga Konsumen Indonesia, rata-rata pemilik 
# mobil harus mengeluarkan sebanyak Rp 170.000,- untuk membeli bensin premium per minggunya.
# Sebanyak 100 sampel yang diambil dari seluruh kota besar di Indonesia menunjukkan bahwa
# rata-rata pengeluaran bensin premium per minggunya adalah Rp 170.145,- dengan standar
# deviasi 500. Apakah kita dapat menyimpulkan bahwa pengeluaran pemilik mobil di Indonesia
# lebih besar dari Rp 170.000,- per minggunya dengan tingkat signifikansi 2,5%? (25 poin)

# H0 : μ = 170000
# H1 : μ > 170000

# α = 0.025
# n = 100
# σ = 500
# x̄ = 170145
# μ = 170000

# Diketahui beberapa nilai yang diperlukan adalah sebagai berikut
n = 100 # jumlah sampel
σp = 500 # simpangan baku populasi
x̄ = 170145 # rata-rata sampel
μ = 170000 # rata-rata populasi
α = 0.025 # tingkat signifikansi

# Menghitung z-score karena n > 30
z = (x̄ - μ) / (σp / np.sqrt(n))
print("z-score:", z)

# Menghitung z-critical
z_critical = stats.norm.ppf(1 - α)
print("z-critical:", z_critical)

# Menentukan keputusan
if z > z_critical:
    print("Tolak H0")
else:
    print("Gagal tolak H0")

# Menghitung p-value dengan one-tailed test karena H1: μ > 170000
p = 1 - stats.norm.cdf(z)
print("p-value:", p)

print("\n")

# SOAL 4
# Sebuah kedai kopitiam di Malang mengklaim bahwa waktu yang dibutuhkan konsumen untuk
# membeli, didistribusikan secara normal dengan rata-rata waktu tunggu 6 menit dengan 
# standar deviasi 1 menit. Untuk membuktikan kebenarannya, tim jaminan kualitas layanan
# dari kopitiam tersebut melakukan penilitian terhadap 81 konsumennya dan diketahui
# rata-rata waktu tunggunya dalah 5 menit. Pada tingkat signifikansi 2%, apakah kita
# bisa menyimpulkan rata-rata waktu tunggu kurang dari 6 menit? (25 poin)

# H0 : μ = 6
# H1 : μ < 6

# α = 0.02
# n = 81
# σ = 1
# x̄ = 5
# μ = 6

# Diketahui beberapa nilai yang diperlukan adalah sebagai berikut
n = 81 # jumlah sampel
σp = 1 # simpangan baku populasi
x̄ = 5 # rata-rata sampel
μ = 6 # rata-rata populasi
α = 0.02 # tingkat signifikansi

# Menghitung z-score karena n > 30
z = (x̄ - μ) / (σp / np.sqrt(n))
print("z-score:", z)

# Menghitung z-critical
z_critical = stats.norm.ppf(α)
print("z-critical:", z_critical)

# Menentukan keputusan
if z < z_critical:
    print("Tolak H0")
else:
    print("Gagal tolak H0")

# Menghitung p-value dengan one-tailed test karena H1: μ < 6
p = stats.norm.cdf(z)
print("p-value:", p)

print("\n")
```
